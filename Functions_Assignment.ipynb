Ans: The difference between a function and a method in Python comes down to how they are used and where they belong:
 Function:
A function is a block of reusable code that is defined using def and can be called independently.

It is not bound to any object.


It is called using dot notation (e.g., object.method()).

Example:



```
# This is fname = "jeet"
```

 2. Explain the concept of function arguments and parameters in Python.

Ans:  Parameters:
Parameters are the placeholders in a function definition.

They act like variables that accept values when the function is called.

You define them inside the parentheses of the function.


```
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

```

Arguments:
Arguments are the actual values you pass to the function when calling it.

These values replace the parameters during execution.



```
greet("Jeet")  # "Jeet" is an argument
```

 4. What is the purpose of the `return` statement in a Python function?

Ans: Purpose of return:
Send a value back to the caller.

Exit the function immediately.

Allows the function’s result to be used or stored elsewhere.




 5. What are iterators in Python and how do they differ from iterables?

Ans: Iterables:
An iterable is any object you can loop over using a for loop.

Examples: list, tuple, string, set, dict, etc.

It has a built-in method called __iter__() which returns an iterator.

```
my_list = [1, 2, 3]
for item in my_list:
    print(item)

```
Iterators:
An iterator is the actual object that does the iteration.

It is created from an iterable using iter().

It keeps track of where it is in the sequence and uses the __next__() method to get the next item.

```
my_list = [1, 2, 3]
iterator = iter(my_list)

print(next(iterator))  # 1
print(next(iterator))  # 2
print(next(iterator))  # 3
# print(next(iterator))  # Error: StopIteration

```





 6. Explain the concept of generators in Python and how they are defined.

 Ans: What is a Generator?
A generator is a special type of function that yields values one at a time, instead of returning all at once.

It allows lazy evaluation, meaning it only computes values when needed, not all at once like a list.

How is a Generator Defined?
Using the yield keyword instead of return:

```
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
      

```
This function doesn't return a single value—it gives you a generator object.





 7. What are the advantages of using generators over regular functions?

Ans: Advantages of Generators Over Regular Functions:

1. Memory Efficiency
Generators don’t store all values in memory.

They generate one value at a time, on the fly.


```
def get_numbers():
    return [i for i in range(1000000)]  # Uses a lot of memory

```

2. Faster Startup Time
A regular function with a huge list waits till the entire list is ready.

A generator starts producing values immediately.

3. Infinite Sequences Possible
Generators can go on forever. Useful for streams, logs, or Fibonacci-like stuff.

 4. Clean & Readable Code
You avoid manually tracking state like in custom iterator classes.


5. Pipelines & Lazy Evaluation
Combine multiple generators to form a data pipeline.

Only process what's needed, when it's needed.

```
def even_numbers(nums):
    for n in nums:
        if n % 2 == 0:
            yield n

```

 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans: 1. map()
Purpose:
* Applies a function to each item in an iterable and returns a new iterator

```
nums = [1, 2, 3, 4]
squared = map(lambda x: x*x, nums)
print(list(squared))  # [1, 4, 9, 16]

```
2. filter()
Purpose:
* Filters out items based on a condition (returns only True ones).



```
nums = [1, 2, 3, 4, 5]
even = filter(lambda x: x % 2 == 0, nums)
print(list(even))  # [2, 4]

```
3. reduce()
Purpose:
* Repeatedly applies a function to the pairs of items and reduces the iterable to a single value.

To use reduce(), you must import it from functools.



```
from functools import reduce

nums = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, nums)
print(total)  # 10

```

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_evens(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example Usage:
my_list = [1, 2, 3, 4, 5, 6]
result = sum_of_evens(my_list)
print("Sum of even numbers:", result)  # Output: 12

````

````python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2 or n > 200:
        return False  # Only checking from 1 to 200
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
````

````python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    """
    Reads a file line by line and yields each line as a string.

    Args:
    file_path (str): The path to the file to read.

    Yields:
    str: Each line of the file as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # strip() removes leading/trailing whitespace, including newline characters
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return  # Exit the generator function if the file is not found

# Example usage:
# Replace 'your_file.txt' with the actual path to your file.
file_path = 'your_file.txt'

for line in read_file_line_by_line(file_path):
    print(line)
````

````python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Conversion formula: F = C * 9/5 + 32
fahrenheit_temps = list(map(lambda c: c * 9/5 + 32, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
````

````python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
input_str = "Hello, World!"

# Use filter to remove vowels
result = ''.join(filter(is_not_vowel, input_str))

print("Original String:", input_str)
print("Without Vowels:", result)
````
